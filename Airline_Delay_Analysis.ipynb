```python
# new column for dep hour. It's in military time so 5:00AM would be listed as 500. Need to divide by 100 and truncate to the hour i.e. 15:45 is going to be 15 for the 15th hour of the day.
df['DEP_HOUR'] = df['SCHEDULED_DEPARTURE'] // 100
```

```python
# departure hour will be the row header and DEPARTURE_DELAY will be the column with the avg of the delay at each hour
hourly_delay = df.groupby('DEP_HOUR')['DEPARTURE_DELAY'].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
hourly_delay.plot(kind='bar')
plt.title("Average Departure Delay by Hour of Day")
plt.xlabel("Scheduled Departure Hour")
plt.ylabel("Average Delay (minutes)")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# make routes into a 'ATL ➝ LGA' format. then count the number of times it was canceled == 1. Then group by the route highest to lowest 

df['ROUTE'] = df['ORIGIN_AIRPORT'].astype(str) + " ➝ " + df['DESTINATION_AIRPORT'].astype(str)
cancelled_routes = (
    df[df['CANCELLED'] == 1]
    .groupby('ROUTE')
    .size()
    .sort_values(ascending=False)
)
```

```python
top_cancelled_routes = cancelled_routes.head(15)

plt.figure(figsize=(10, 6))
top_cancelled_routes.plot(kind='barh', color='darkred')
plt.title("Top 15 Most Cancelled Routes (2015)")
plt.xlabel("Number of Cancellations")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
route_total = df_clean.groupby('ROUTE').size()

# Total cancellations per route
route_cancellations = df_clean[df_clean['CANCELLED'] == 1].groupby('ROUTE').size()

# Combine into a DataFrame
cancel_rate_df = pd.DataFrame({
    'Total_Flights': route_total,
    'Cancellations': route_cancellations
})

# Fill NaNs with 0 cancellations
cancel_rate_df['Cancellations'] = cancel_rate_df['Cancellations'].fillna(0)

# Calculate rate
cancel_rate_df['Cancellation_Rate'] = (cancel_rate_df['Cancellations'] / cancel_rate_df['Total_Flights']).round(4)
```

```python
cancel_rate_df_filtered = cancel_rate_df[cancel_rate_df['Total_Flights'] >= 50]
```

```python
top_cancel_rate_routes = (
    cancel_rate_df_filtered
    .sort_values(by='Cancellation_Rate', ascending=False)
    .head(15)
)
```

```python
import matplotlib.pyplot as plt

# plot routes by cancellation rate. 0.10 represents 10%

plt.figure(figsize=(10, 6))
top_cancel_rate_routes['Cancellation_Rate'].plot(kind='barh', color='crimson')
plt.title("Top 15 Routes by Cancellation Rate (2015)")
plt.xlabel("Cancellation Rate")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
